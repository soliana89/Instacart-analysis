# Q.4 Create a loyalty flag for existing customers

```python
#split data into groups based on user_id, and tranform order_number to generate max orders for each user
ords_prods_merged['max_order']= ords_prods_merged.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
#check accuracy
pd.options.display.max_rows= None
ords_prods_merged.head(100)
```

```python
#derive column conditions
ords_prods_merged.loc[ords_prods_merged['max_order']>40, 'loyalty_flag']= 'Loyal customer'
```

```python
ords_prods_merged.loc[(ords_prods_merged['max_order']<=40) & (ords_prods_merged['max_order']>10), 'loyalty_flag']= 'Regular customer'
```

```python
ords_prods_merged.loc[ords_prods_merged['max_order']<=10, 'loyalty_flag']= 'New customer'
```

```python
#frequency of loyalty_flag
ords_prods_merged['loyalty_flag'].value_counts(dropna=False)
```

```python
ords_prods_merged[['user_id', 'loyalty_flag', 'order_number']].head(20)
```

# Q.5 Does the price of products purchased by loyal customers differ from regular or new customers


```python
ords_prods_merged.groupby('loyalty_flag').agg({'prices':['mean', 'min', 'max', 'sum']})
```

Loyal customers on average spend the least amount of money on their products, whereas new customers spend the most amount of money on their purchases.Â¶

# Q.6 split data into groups based on user_id, and transform prices to generate average_price for each user

```python
ords_prods_merged['average_price']= ords_prods_merged.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
#check accuracy
ords_prods_merged.head(60)
```

```python
#derive column conditions
ords_prods_merged.loc[ords_prods_merged['average_price']<10, 'spending_flag']= 'Low spender'
```

```python
ords_prods_merged.loc[ords_prods_merged['average_price']>=10, 'spending_flag']= 'High spender'
```

```python
#frequency of spending_flag
ords_prods_merged['spending_flag'].value_counts(dropna= False)
```

```python
ords_prods_merged[['user_id', 'spending_flag', 'loyalty_flag', 'order_number']].head(20)
```

# Q.7 split data into groups based on user_ids, and transform days_since_prior_order to generate median_order_frequency

```python
ords_prods_merged['median_order_frequency']= ords_prods_merged.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
#check accuracy
ords_prods_merged.head(60)
```

```python
#derive column conditions
ords_prods_merged.loc[ords_prods_merged['median_order_frequency']>20, 'order_frequency_flag']= 'Non-frequent customer'
```

```python
ords_prods_merged.loc[(ords_prods_merged['median_order_frequency']>10) & (ords_prods_merged['median_order_frequency']<=20), 'order_frequency_flag']= 'Regular customer'
```

```python
ords_prods_merged.loc[ords_prods_merged['median_order_frequency']<=10, 'order_frequency_flag']= 'Frequent customer'
```

```python
#frequency of order_frequency_flag
ords_prods_merged['order_frequency_flag'].value_counts(dropna= False)
```

```python
ords_prods_merged[['user_id', 'loyalty_flag', 'spending_flag', 'order_frequency_flag']].head(60)
```
